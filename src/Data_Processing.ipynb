```python
prj_dir = 'G:\\DATA_ANALYTICS_PROJECT\\ESPN\\data-analytics-project-for-beginners\\t20_json_files'
csv_path= 'G:\\DATA_ANALYTICS_PROJECT\\ESPN\\data-analytics-project-for-beginners\\t20_csv_files'
match_results_path = f"{prj_dir}\\t20_wc_match_results.json"
batting_summary_path = f"{prj_dir}\\t20_wc_batting_summary.json"
bowling_summary_path = f"{prj_dir}\\t20_wc_bowling_summary.json"
player_info = f"{prj_dir}\\t20_wc_player_info.json" 
```

```python
with open(match_results_path) as f:
    data = json.load(f)
data[0]['matchSummary']
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard': 'match_id'}, axis=1, inplace=True)
```

```python
df_match.head()
```

# BATTING SUMMARY

```python
with open(batting_summary_path) as b:
    data = json.load(b)
    list = []
    for i in data:
        list.extend(i["battingSummary"])
    list

df_bat = pd.DataFrame(list)
df_bat.head()
df = df_bat.copy()
```

```python
df_bat["out/notout"] = df_bat.dismissal.apply(lambda x: "out" if len(x)>0 else "notout")
df_bat.drop("dismissal", axis=1, inplace = True)
```

```python
df_bat.head()
```

```python
df_bat["batsmanName"] = df_bat.batsmanName.apply(lambda x: x.replace("â€", ""))
df_bat["batsmanName"] = df_bat.batsmanName.apply(lambda x: x.replace("\xa0", ""))
df_bat.head()
```

```python
match_ids_dict = {}
for index, row in df_match.iterrows():
    key1 = row["team1"] + " Vs " + row["team2"]
    key2 = row["team2"] + " Vs " + row["team1"]
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict
```

```python
df_bat["match_id"] = df_bat["match"].map(match_ids_dict)
df_bat.head()
```

# BOWLING SUMMARY

```python
with open(bowling_summary_path) as b:
    data = json.load(b)
    list = []
    for i in data:
        list.extend(i["bowlingSummary"])
    list

df_bowl = pd.DataFrame(list)
df_bowl.head()
```

```python
df_bowl["match_id"] = df_bowl["match"].map(match_ids_dict)
df_bowl.head()
```

# PLAYER INFORMATION 

```python
with open(player_info) as p:
    data = json.load(p)
df_player = pd.DataFrame(data)
df_player.head()
```

```python
df_player["name"] = df_player["name"].apply(lambda x: x.replace("â€", ""))
df_player["name"] = df_player["name"].apply(lambda x: x.replace('†', ""))
df_player["name"] = df_player["name"].apply(lambda x: x.replace('\xa0', ""))
df_player.head()
```

```python
df_bat.to_csv(f"{csv_path}\\t20_wc_batting_summary.csv", index = False)
df_bowl.to_csv(f"{csv_path}\\t20_wc_bowling_summary.csv", index = False)
df_player.to_csv(f"{csv_path}\\t20_wc_player_info.csv", index = False)
```
